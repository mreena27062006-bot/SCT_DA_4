```python
# Check Unique Values in Text Columns
print("Education values:", df['Education'].unique())
print("Marital Status values:", df['Marital_Status'].unique())
```

```python
# Convert Dt_Customer from text to proper date format
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer']).dt.strftime('%Y-%m-%d')
print("✅ Date column fixed!")
print(df['Dt_Customer'].dtype)  # Shows : datetime64
```

```python
# Customer Age (from birth year)
df['Age'] = 2025 - df['Year_Birth']

# How many years they've been a customer
df['Years_As_Customer'] = (pd.Timestamp('2025-01-01') - df['Dt_Customer']).dt.days // 365

# Total money spent across ALL product categories
df['Total_Spent'] = (df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] +
                     df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds'])

# Total number of purchases across ALL channels
df['Total_Purchases'] = (df['NumWebPurchases'] + df['NumCatalogPurchases'] +
                         df['NumStorePurchases'] + df['NumDealsPurchases'])

# Total campaigns accepted
df['Total_Campaigns_Accepted'] = (df['AcceptedCmp1'] + df['AcceptedCmp2'] +
                                   df['AcceptedCmp3'] + df['AcceptedCmp4'] +
                                   df['AcceptedCmp5'] + df['Response'])

print("✅ New columns created!")
print(df[['Age', 'Years_As_Customer', 'Total_Spent', 
          'Total_Purchases', 'Total_Campaigns_Accepted']].head())
```

```python
# Fill the 24 missing Income values with the median income
# Median is better than average here because of the outlier (666,666)

print("Missing Income before:", df['Income'].isnull().sum())

# ✅ Updated syntax — works perfectly in all pandas versions
df['Income'] = df['Income'].fillna(df['Income'].median())

print("Missing Income after:", df['Income'].isnull().sum())  # Should show 0
```

```python
print("Rows before removing outliers:", len(df))

# Remove impossible ages (born before 1900 = age > 125)
df = df[df['Age'] <= 100]

# Remove the Income outlier (666,666 is clearly a data error)
df = df[df['Income'] < 200000]

print("Rows after removing outliers:", len(df))
# We will lose only about 4-5 rows — that's normal and correct
```

```python
# Group customers into segments based on income
# This helps Tableau create the "Spending by Segment" chart

# Better segment boundaries for this dataset
df['Income_Segment'] = pd.cut(
    df['Income'],
    bins=[0, 35000, 55000, 75000, 200000],
    labels=['Low', 'Mid', 'High', 'Premium']
)
print(df['Income_Segment'].value_counts())
```

```python
print("=== FINAL CLEANED DATASET ===")
print("Shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
print("\nNew columns added:", ['Age', 'Years_As_Customer', 'Total_Spent', 
                               'Total_Purchases', 'Total_Campaigns_Accepted', 
                               'Income_Segment'])
```

```python
df.to_csv('marketing_campaign_cleaned.csv', index=False)
print("✅ marketing_campaign_cleaned.csv is ready for Tableau!")
print("Rows:", len(df), "| Columns:", len(df.columns))
```
